# Part II - PCA for Handwritten Digit Decomposition (35 marks)

In this part, we will analyze a dataset of handwritten digits using principal component analysis (PCA). We have learned "eigenfaces" in class, now we compute "eigendigits" to understand how principal components capture variations in digit shapes.

This utility function will just let us convert a vector into an image, so it can be shown on the screen.

Now, we want to randomly pick **1000** images to compute their eigendigits, i.e., the directions of the principal components for this collection of images.
                                          
When using PCA, we should do all the processing for the **centred** data. Therefore, take the following steps:
1. Define **<font color='red'>mean_digit</font>** to be the mean of all images of handwritten digits for the 1000 random images.
1. Subtract **mean_digit** from all the 1000 images. Place the result in **<font color='red'>X_centered</font>**.
1. Find the left-singular vectors of **X_centered** (i.e., the matrix U in the SVD) and place the result in **<font color='red'>U</font>**. Find the right-singular vectors of **X_centered** (i.e., the matrix V in the SVD) and place the result in **<font color='red'>V</font>**, with **<font color='red'>Vt</font>** represent the transpose of V.

NOTE: At the end of this process, the columns of **Vt** should represent the **eigendigits** due to the data shape - the dataset stores the number of images and then each image is a 28*28 vector.<br>

**<font color=blue>IMORTANT:</font>** When using **svd** (from np.linalg), it is important to specify **full_matrices=False**. Otherwise, the algorithm aims to compute the full dimensional matrices, which are not needed and require a lot of memory.
                       

````python
# Randomly select 1000 images
np.random.seed(42)  # For reproducibility
random_indices = np.random.choice(nimages, 1000, replace=False)
digits_subset = digits[random_indices]

#Reshape images and compute mean digit
# Reshape each 28x28 image into a 784-dimensional vector
X = digits_subset.reshape(1000, -1)  # Shape: 1000 x 784

# Computing the mean digit
mean_digit = X.mean(axis=0)  # Shape: (784,)

# Center the data
X_centered = X - mean_digit  # Shape: 1000 x 784

#Step 4: Compute SVD (with full_matrices=False for efficiency)
U, s, Vt = np.linalg.svd(X_centered, full_matrices=False)

# The principal components (eigendigits) are the rows of Vt
eigendigits = Vt  # Shape: 784 x 784 (but we only use the first few)

# Verificatio n of shapes
print("U shape:", U.shape)      # Should be (1000, 784)
print("s shape:", s.shape)      # Should be (784,)
print("Vt shape:", Vt.shape)    # Should be (784, 784)

# Visualising the mean digit and first few eigendigits
plt.figure(figsize=(12, 6))

# Plot mean digit
plt.subplot(2, 5, 1)
plt.imshow(vec2img(mean_digit), cmap='gray')
plt.title('Mean Digit')
plt.axis('off')

# Plot first 9 eigendigits (principal components)
for i in range(9):
    plt.subplot(2, 5, i+2)
    plt.imshow(vec2img(Vt[i]), cmap='gray')
    plt.title(f'Eigendigit {i+1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
````

Another part is determining how many eigendigits are needed for a reasonable reconstruction of the original images. For this, plotting the cumulative explained variance against the number of components would be useful. The explained variance can be calculated from the singular values stored in S_all. The cumulative sum of squared singular values divided by the total sum gives the proportion of variance explained. This graph will show how adding more components captures more variance, helping to decide the optimal number. We usually say 90% is a very good reconstruction for images. How many eigendigits does it require to reach 90%?

It reaches 90% at k = 79

````python
# Calculating explained variance from singular values (s)
total_variance = np.sum(s**2)
explained_variance = (s**2) / total_variance
cumulative_variance = np.cumsum(explained_variance)

# Finding where we reach 90% variance
k_90 = np.argmax(cumulative_variance >= 0.9) + 1

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(cumulative_variance, 'b-', linewidth=2)
plt.axhline(y=0.9, color='r', linestyle='--')
plt.axvline(x=k_90, color='r', linestyle='--')
plt.xlabel('Number of Eigendigits (Principal Components)')
plt.ylabel('Cumulative Explained Variance')
plt.title('Variance Explained by Eigendigits')
plt.grid(True)

# Annotating the 90% point
plt.text(k_90+5, 0.85, f'90% at {k_90} components', 
         color='red', fontsize=12)
plt.text(5, 0.92, '90% threshold', 
         color='red', fontsize=10)

# Adding some reference markers
for k in [10, 50, 100]:
    y_val = cumulative_variance[k-1]
    plt.plot(k, y_val, 'ko')
    plt.text(k+2, y_val-0.03, f'{y_val:.1%} at k={k}', fontsize=9)

plt.show()

print(f"Number of eigendigits needed for 90% variance: {k_90}")
````

# Part III -  Matrix Completion for Image Recovery (30 marks)

In this part, we perform  matrix completion  using the **Singular Value Thresholding** (SVT) algorithm we saw in class.<br>
Given an incomplete matrix **M**, recall that our objective is to solve:
$$ X^* = \arg\min_{X} \tau\|X|_* + \frac{1}{2} \|X\|_F^2\quad\text{subject to}\quad P_{\Omega}(X)=P_\Omega(M),$$
where $\Omega$ represents the set of indexes where the values of **M** are **known**.

The SVT algorithm should have the following update iterations:
$$
\begin{split}
X^{(k+1)} &= D_\tau(Y^{(k)}),\\
Y^{(k+1)} &= Y^{(k)} + \delta P_\Omega(M-X^{(k+1)}).\end{split}
$$
The starting value should be $Y^{(0)} = 0$.<br> 
The algorithm should stop after $k$ steps if:
$$ \frac{\|P_\Omega(M)-P_\Omega(X^{(k)})\|_F}{\|P_\Omega(M)\|_F} \le \text{tolerance}.$$


Place your code in the function **<font color='red'>matrix_complete</font>**. It should take the following input:
* **<font color='red'>M</font>** - the incomplete matrix in $\mathbb{R}^{D\times N}$. The **known** values are assumed to be between 0 and 1, and an unknown value will be marked by **<font color='red'>-1</font>**.
* **<font color='red'>tolerance</font>** - the maximum tolerance to decide when iterations should stop. The default value is $1.0\times 10^{-5}$.
* **<font color='red'>delta_val</font>** - a value for the increment size $\delta$.<br>
If **delta_val** is not specified (=None), the default value should be $\delta = 0.80\times\frac{D\times N}{K}$, where **K** is the number of **known** entries in **M**.
* **<font color='red'>tau_val</font>** - a value for the shrinking size $\tau$. <br>
If **tau_val** is not specified (=None), the default value should be $\tau = 0.40\times\max(D,N)$.
* **<font color='red'>max_iter</font>** - maximum number of iterations to run.
* **<font color='red'>iter_print</font>** - how often to print out a progress line.

The output of the function should be the final value of **<font color='red'>X</font>**.

````python
def svt_operator(Y, tau):
    U, S, Vt = np.linalg.svd(Y, full_matrices=False)
    S_thresholded = np.maximum(S - tau, 0)
    return U @ np.diag(S_thresholded) @ Vt
````

````python
import numpy as np
from numpy.testing import assert_almost_equal

def matrix_complete(M, tolerance=1e-5, delta_val=None, tau_val=None, max_iter=500, iter_print=50):
    M = np.array(M)
    D, N = M.shape
    known_mask = M != -1
    M_known = np.where(known_mask, M, 0)  # Replace unknowns with 0 for masking

    K = np.sum(known_mask)
    if delta_val is None:
        delta_val = 0.8 * D * N / K
    if tau_val is None:
        tau_val = 0.4 * max(D, N)

    Y = np.zeros((D, N))
    for k in range(max_iter):
        X = svt_operator(Y, tau_val)
        residual = np.where(known_mask, M_known - X, 0)
        Y = Y + delta_val * residual

        error = np.linalg.norm(residual, 'fro') / np.linalg.norm(M_known, 'fro')
        if k % iter_print == 0:
            print(f"Iter {k}: relative error = {error:.6f}")
        if error < tolerance:
            break

    return X
````

The noise can appear in any RGB channels independently. If the image is of width **W** and height **H**, we will process it as a matrix with **H rows** and **3W columns**.<br>
We start by writing functions to convert between the two formats.

We first need to code two functions:
* **<font color=red>img2mat</font>** - converts an image from an **(H x W x 3)** 3d-array to an **(H x 3W)** matrix.
* **<font color=red>mat2img</font>** - converts an **(H x 3W)** matrix back into an **(H x W x 3)** 3d-array.

````python
def img2mat(image):
    H, W, C = image.shape  # C = 3 for RGB
    return image.reshape(H, W * C)

def mat2img(mat):
    H = mat.shape[0]
    W = mat.shape[1] // 3  # 3 channels 
    return mat.reshape(H, W, 3)
````

Now we set up an **<font color='red'>image_recovery</font>** function that takes in the noisyimage and the original image, within which we will apply the complete_matrix function defined earlier to recover the **noisyimage**. We first need to convert the **noisyimage** and the original image into matrices, namingly **<font color='red'>noisy_matrix</font>** and **<font color='red'>original_matrix</font>** using the **img2mat** function. 

After that, we apply the matrix completion and store the recovered image matrix in variable **<font color='red'>recovered_matrix</font>**. Set tolerance=1e-4, $\tau = 0.4*max(D,N)$ and $\delta = 0.80\times\frac{1.0}{0.3}$, the total number of pixels over the number of known ones since 70% pixels are noise. We convert **recovered_matrix** back to image, and store it in **<font color='red'>recovered</font>**.

We return the recovered, the noisy, and the original images from this **image_recovery** function.

````python
def image_recovery(noisyimage, original_image):
    # Converting both image to matrix form 
    noisy_matrix = img2mat(noisyimage)
    original_matrix = img2mat(original_image)

    # Set parameters now
    tol = 1e-4
    D, N = noisy_matrix.shape
    delta = 0.80 * (1.0 / 0.3)  # Since 70% are missing (30% known)
    tau = 0.4 * max(D, N)

    # Recovering the mtarix using matrix completion
    recovered_matrix = matrix_complete(noisy_matrix, tolerance=tol, delta_val=delta, tau_val=tau)

    # Converting back to image format
    recovered = mat2img(recovered_matrix)

    return recovered, noisyimage, original_image
````

In the next step, we recover the noisy image by calling the **image_recovery** function. This process will take around 5 minutes.

````python
recovered, noisy, original = image_recovery(noisyimage, image)
````
